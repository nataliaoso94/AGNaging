```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = open('Catalogo_Muestra_oficial.tsv', 'r')
lines = data.readlines()   	
data.close()
objname = []
logLMir = []
logLX = []
logLOIII = []


for line in lines:
    for mm in np.arange(100):
        line = line.replace("  "," ")
    if not line.startswith('#'):
        line = line.replace('\n', ' ')
        line = line.replace('\s', ' ')
        line = line.replace('\t', ' ')
        columns = line.split(" ")
        objname.append(str(columns[5]))
        logLOIII.append(float(columns[6]))
        logLMir.append(float(columns[7]))
        logLX.append(float(columns[8]))
logLOIII = np.array(logLOIII)
logLMir = np.array(logLMir)
logLX = np.array(logLX)
```

```python
lxlmir =(logLMir +4.37)/1.106 
lmirloiii = 1.018*logLOIII + 2.56
lxloiii =  0.92*logLOIII+6.27
```

```python
fig = plt.figure(figsize = (8,8))
plt.plot(logLMir,logLX, 'o', color = 'green')
plt.ylabel(r'$\rm{log \ L_{2-10 kev}}$', fontsize = 20)
plt.xlabel(r'$\rm{log \ \nu L_{12\mu}}$', fontsize = 20)
plt.plot(logLMir, lxlmir, color = 'green', label = 'LMir LX relation')
```

```python
fig = plt.figure(figsize = (8,8))
plt.plot(logLOIII,logLMir, 'o', color = 'blue')
plt.xlabel(r'$\rm{log \ \nu L_{12\mu}}$', fontsize = 20)
plt.ylabel(r'$\rm{log \ L_{OIII}}$', fontsize = 20)
plt.plot(logLOIII, lmirloiii, color = 'blue', label = 'LMir LOIII relation')
```

```python
fig = plt.figure(figsize = (8,8))
plt.plot(logLOIII, logLX,'o', color = 'red')
plt.ylabel(r'$\rm{log \ L_{2-10 kev}}$', fontsize = 20)
plt.xlabel(r'$\rm{log \ L_{OIII}}$', fontsize = 20)
plt.plot(logLOIII, lxloiii, color = 'red', label = 'LOIII LX relation')
```
